#**Inspecting Data**

When you find a data, you have to inspect it at first, kinda laking looking around to get some basic informations like shape, element number, whether there is a missing or null value, the missing values' locations, and many more. Let's start!

```python
print(housing_df.shape)
print(housing_df.head())
print(housing_df.tail())
print(housing_df.info())
```

The shape will give the row and column number, head and tail will give first 5 and last 5 rows, respectively, and info will give a general information. Here entries means total element number (20640 entries), non-null means not missing or existing values, dtype refers to data type.

##**Missing Value**
The missing value part is important, and you have to look out for missing values in any database while preparing it for any use. There are more ways to deal about it!

```python
housing_df.isnull().sum()
```

```python
dia_df.isnull().sum()
```

There are no missing values there it seems.



##**Couting similars**

There is a value_count method which can be used to know about the appearance frequency of any numbers. For example, let's say you want to solve a murder mystery. Then you will look for frequencies of the numbers that called the victim. This is something similar! Let's look in the diabetes dataset.

```python
dia_df.head()
```

```python
dia_df.tail()
```

From them, Pregnancy, Outcome, Age might be good to check for the purpose of checking data bias.

```python
dia_df.value_counts(['Outcome'])
```

```python
dia_df.value_counts(['Pregnancies'])
```

```python
dia_df.value_counts(['Age'])
```

##**Mean function along with grouping!**

This is a nice one. You will know its power by seeing it

```python
dia_df.groupby(['Outcome']).mean()
```

```python
dia_df.groupby(['Pregnancies']).mean()
```

##**Manipulating DataFrame**
